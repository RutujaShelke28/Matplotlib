```python
import matplotlib.pyplot as plt
```

```python
subjects = ["maths", "science", "social studies", "geography", "drawing"]
time = [4,3,1,2,5]
colors = ["skyblue", "pink", "purple", "navy", "yellow"]

plt.figure(figsize=(6,6))
plt.pie(time, labels = subjects, colors = colors, autopct = '%1.1f%%')
```

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5,6,7]
dsa = [6,3,4,2,5,1,7]
lindkin = [2,4,1,3,5,6,2]
github = [1,6,3,4,7,2,5]

labels = ['days', 'dsa', 'lindkin', 'github']
colors = ['#DEB6C5', '#5C5056', '#C9B6E0', '#B6E0BA']
##plt.figure(figsize=(7,6))
plt.stackplot(days, dsa, lindkin, github, labels = labels, colors = colors, alpha = 0.7)
plt.legend(loc = "upper left")
plt.title("MY DAILY ACTIVITES")
plt.xlabel("Days")
plt.ylabel("Activities")
```
