```python
import matplotlib.pyplot as plt
```

```python
study_hrs = [1,2,3,4,5,6]
marks = [20,30,40,50,60,70]
colors = ["red" if marks < 40 else "green" for marks in study_hrs]

plt.scatter(study_hrs, marks, c = colors)
plt.grid(True)
```

```python
# Size of points based on score (bigger score -> bigger point)
study_hours = [1,2,3,4,5,6]
exam_scores = [20,30,40,50,60,70]
sizes = [score * 2 for score in exam_scores]
colors = ['red' if score < 60 else 'green' for score in exam_scores]

plt.scatter(study_hours, exam_scores, s=sizes, c=colors)
plt.title('Colored & Sized Scatter Plot')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
study_hours = [1,2,3,4,5,6]
exam_scores = [20,30,40,50,60,70]
sizes = [score * 2 for score in exam_scores]
##colors = ['red' if score < 60 else 'green' for score in exam_scores]

plt.scatter(study_hours, exam_scores, s=sizes, c=exam_scores, cmap = 'viridis')
plt.colorbar(label = "Exam Scores")
plt.title('Colored & Sized Scatter Plot')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
plt.scatter(study_hours, exam_scores)

# Add labels
for i in range(len(study_hours)):
    plt.annotate(f'S {i+1}', (study_hours[i], exam_scores[i]))

plt.title('Scatter Plot with Annotations')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.colorbar(label = "Study Hours")
plt.grid(True)
plt.show()
```
